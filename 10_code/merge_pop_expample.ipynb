Census data for WA

```python
import numpy as np
import pandas as pd
```

```python
pop_2000_2009 = pd.read_csv('https://www.dropbox.com/s/5ah76tcq8r9gy60/Washington%20Census%20Data%202000%20-%202009.csv?dl=1', header=3)
pop_2010_2019 = pd.read_csv('https://www.dropbox.com/s/g2tnmlubdq8592e/Washington%20Census%20Data%202010%20-%202019.csv?dl=1', header=3)
```

```python
pop_2000_2009_ = pop_2000_2009.iloc[1:40, :12].drop(['Unnamed: 1'], axis = 1)
pop_2000_2009_.rename(columns={'Unnamed: 0': 'COUNTY'}, inplace=True)
pop_2000_2009_.head()
```

```python
pop_2010_2019_ = pop_2010_2019.iloc[1:40, :].drop(['Census', 'Estimates Base'], axis = 1)
pop_2010_2019_.head()
```

```python
# Unnamed: 0 is county, state, split it into county and state separately
pop_2010_2019_[['COUNTY', 'STATE']] = pop_2010_2019_['Unnamed: 0'].str.split(',', 1, expand=True)

# drop the original column
pop_2010_2019_.drop(['Unnamed: 0', 'STATE'], axis = 1, inplace = True)

pop_2010_2019_.head()
```

```python
assert pop_2000_2009_.shape[0] == pop_2010_2019_.shape[0]  # number of county should be the same
```

```python
pop_2000_2019 = pd.merge(pop_2000_2009_, pop_2010_2019_, how = 'outer', on = 'COUNTY')
pop_2000_2019.head()
```

```python
# convert columns to rows
pop_WA = pd.melt(pop_2000_2019, id_vars=['COUNTY'], var_name='YEAR', value_name='POPULATION')
pop_WA.head()
```

```python
assert pop_WA.shape[0] == pop_2000_2019.shape[0] * (pop_2000_2019.shape[1]-1)  # number of rows should be the same
assert pop_WA['YEAR'].nunique() == (pop_2000_2019.shape[1]-1)  # number of years should be the same
```

```python
# remove the dot in the county name
pop_WA['COUNTY'] = pop_WA['COUNTY'].str.replace('.', '').str.strip()
pop_WA.head()
```

```python
# Add a column for state
pop_WA['STATE'] = 'WA'
pop_WA.head()
```
