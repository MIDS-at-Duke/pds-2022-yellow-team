### Shipment: Florida and its reference states

```python
fl_ship_pop = pd.read_csv('../20_intermediate_files/fl_shipment_pop.csv')
fl_ship_pop.head()
```

```python
all_death = pd.read_csv('../20_intermediate_files/only_od_all_years.csv')
all_death.head()
```

```python
# select only florida and its reference states
fl_ref_states = ['AZ', 'FL', 'LA', 'SC']
fl_ship_death = all_death[all_death['STNAME'].isin(fl_ref_states)]

# rename the abbrevation state name to its full name
fl_ship_death_ = fl_ship_death.replace({'AZ': 'Arizona', 'FL': 'Florida', 
                                        'LA': 'Louisiana', 'SC': 'South Carolina'})

# make sure the state names are the same
assert (fl_ship_death_['STNAME'].unique() == fl_ship_pop['STNAME'].unique()).all()

# rename year
fl_ship_death_ = fl_ship_death_.rename(columns={'Year': 'YEAR'})
```

```python
# merge the two dataframes
fl_ship_death_pop = pd.merge(fl_ship_pop, fl_ship_death_, how='left', on = ['STNAME', 'CTYNAME', 'YEAR'])
fl_ship_death_pop[fl_ship_death_pop['YEAR'] == 2003]
```
